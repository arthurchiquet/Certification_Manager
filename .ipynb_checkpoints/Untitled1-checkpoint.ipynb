```python
import pandas as pd
import json
```

```python
df=pd.read_csv('data/ParcellaireDetail_22_02_2021.csv', sep=';')
```

```python
with open('data/cadastre-51210-parcelles.json') as file:
    geojson=json.load(file)
```

```python
features=geojson['features']
```

```python
db=[{'coordinates':i['geometry']['coordinates'], 'commune':i['properties']['commune'], 'section':i['properties']['section'], 'numero':i['properties']['numero']} for i in features]
```

```python
df=df.rename(columns={'code_commune':'commune',' section cadastrale':'section','numero plan':'numero'})
```

```python
df2=pd.DataFrame.from_dict(db)
```

```python
df2=df2.astype({'commune': 'int64', 'section':'object','numero':'int64'})
```

```python
new_df = df.merge(df2,  how='left', left_on=['commune','section','numero'], right_on = ['commune','section','numero'])
```

```python
new_df
```

```python
new_df=new_df[new_df.commune==51210]
```

```python
new_df.to_dict(orient='records')
```

```python
new_dict=[{'type': 'Feature','geometry': {'type': 'Polygon','coordinates' : i['coordinates']}, 'properties':{'propriétaire':i['nom proprietaire'], 'cépage':i['nom cepage'], 'annee plantation':i['annee plantation'], 'sp': i['surface plantee']}} for i in new_df.to_dict('records')]
```

```python
new_dict
```

```python
new_json={"type":"FeatureCollection","features":new_dict}
```

```python
with open('data/geojson.json', 'w') as file:
    json.dump(new_json, file)
```
